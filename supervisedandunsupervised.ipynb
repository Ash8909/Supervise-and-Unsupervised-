```python
import pandas as pd
import numpy as np

# Number of anime watchers
n_watchers = 75

# Generate watcher names (simple placeholders)
watcher_names = [f'Watcher_{i+1}' for i in range(n_watchers)]

# List of popular anime shows
anime_shows = [
    'Attack on Titan', 'Jujutsu Kaisen', 'Demon Slayer', 'My Hero Academia',
    'One Piece', 'Naruto', 'Dragon Ball Z', 'Death Note', 'Fullmetal Alchemist: Brotherhood',
    'Hunter x Hunter', 'Spirited Away', 'Princess Mononoke', 'Your Name.', 'A Silent Voice'
]

# Generate random anime show selections
watched_shows = np.random.choice(anime_shows, n_watchers)

# Generate random hours spent watching (e.g., between 10 and 300 hours)
hours_watching = np.random.randint(10, 301, n_watchers)

# Create a dictionary from the generated data
anime_data = {
    'Name': watcher_names,
    'Show Watched': watched_shows,
    'Hours Watched': hours_watching
}

# Create the pandas DataFrame
anime_df = pd.DataFrame(anime_data)

# Display the first few rows of the created dataset
display(anime_df.head())

print(f"\nDataset created with {len(anime_df)} anime watchers.")
```

```python
import pandas as pd
import numpy as np

# Number of K-Pop fans
n_fans = 80

# Generate fan names (simple placeholders)
fan_names = [f'Fan_{i+1}' for i in range(n_fans)]

# Generate ages between 15 and 40 (inclusive)
fan_ages = np.random.randint(15, 41, n_fans)

# List of popular K-Pop groups
kpop_groups = [
    'BTS', 'BLACKPINK', 'EXO', 'TWICE', 'Stray Kids',
    'Red Velvet', 'NCT', 'SEVENTEEN', 'ITZY', 'Tomorrow X Together',
    'IVE', 'NewJeans', 'LE SSERAFIM', 'ATEEZ', 'ENHYPEN'
]

# Generate random group support
fan_groups = np.random.choice(kpop_groups, n_fans)

# Create a dictionary from the generated data
kpop_data = {
    'Name': fan_names,
    'Age': fan_ages,
    'Supported Group': fan_groups
}

# Create the pandas DataFrame
kpop_df = pd.DataFrame(kpop_data)

# Display the first few rows of the created dataset
display(kpop_df.head())

print(f"\nDataset created with {len(kpop_df)} K-Pop fans.")
```
