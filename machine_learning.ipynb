```python
import pandas as pd
import numpy as np
import random

# Set a seed for reproducibility
np.random.seed(42)
random.seed(42)

# Define possible values for categorical columns
algorithms = ['Linear Regression', 'Logistic Regression', 'Decision Tree', 'Random Forest', 'Support Vector Machine', 'Gradient Boosting', 'K-Means', 'Neural Network']
task_types = ['Classification', 'Regression', 'Clustering']
dataset_sizes = ['Small', 'Medium', 'Large', 'Very Large']

# Generate 100 rows of data
data = {
    'id': range(1, 101),
    'algorithm': random.choices(algorithms, k=100),
    'task_type': random.choices(task_types, k=100),
    'dataset_size_category': random.choices(dataset_sizes, k=100),
    'num_features': np.random.randint(5, 200, 100),
    'num_samples': np.random.randint(100, 100000, 100),
    'accuracy': np.random.uniform(0.5, 0.99, 100),
    'f1_score': np.random.uniform(0.4, 0.98, 100),
    'precision': np.random.uniform(0.4, 0.98, 100),
    'recall': np.random.uniform(0.4, 0.98, 100),
    'runtime_seconds': np.random.uniform(0.1, 3600, 100), # Runtimes from 0.1 seconds to 1 hour
    'is_optimized': np.random.choice([True, False], 100, p=[0.3, 0.7])
}

df = pd.DataFrame(data)

# Ensure accuracy, f1_score, precision, recall are not greater than 1
df['accuracy'] = np.clip(df['accuracy'], 0, 1)
df['f1_score'] = np.clip(df['f1_score'], 0, 1)
df['precision'] = np.clip(df['precision'], 0, 1)
df['recall'] = np.clip(df['recall'], 0, 1)

# Display the first few rows of the generated DataFrame
print("Generated DataFrame head:")
display(df.head())

# Save the DataFrame to a CSV file
csv_file_path = 'machine_learning_dataset.csv'
df.to_csv(csv_file_path, index=False)
print(f"\nDataset saved to {csv_file_path}")
```

```python
import pandas as pd
import numpy as np
import random

# Set a seed for reproducibility
np.random.seed(42)
random.seed(42)

# Define possible values for categorical columns related to ML experiments
model_architectures = ['CNN', 'RNN', 'Transformer', 'MLP', 'SVM', 'Decision Tree', 'XGBoost', 'LightGBM']
data_preprocessing_methods = ['StandardScaler', 'MinMaxScaler', 'Normalization', 'OneHotEncoding', 'None']
feature_engineering_techniques = ['PCA', 'TSNE', 'Polynomial Features', 'Feature Scaling', 'None']
hyperparameter_tuning_methods = ['Grid Search', 'Random Search', 'Bayesian Optimization', 'Manual']
optimizers = ['Adam', 'SGD', 'RMSprop', 'Adagrad']
dataset_types = ['Image', 'Text', 'Tabular', 'Time Series', 'Graph']
hardware_options = ['GPU', 'CPU', 'TPU']

# Generate 100 rows of data for an unsupervised ML experiment dataset
data_ml_experiments = {
    'experiment_id': range(1, 101),
    'model_architecture': random.choices(model_architectures, k=100),
    'data_preprocessing_method': random.choices(data_preprocessing_methods, k=100),
    'feature_engineering_techniques': random.choices(feature_engineering_techniques, k=100),
    'hyperparameter_tuning_method': random.choices(hyperparameter_tuning_methods, k=100),
    'num_epochs': np.random.randint(5, 200, 100),
    'batch_size': np.random.choice([16, 32, 64, 128], 100),
    'learning_rate': np.random.uniform(0.0001, 0.1, 100),
    'optimizer': random.choices(optimizers, k=100),
    'regularization_strength': np.random.uniform(0.0, 0.1, 100),
    'dataset_type': random.choices(dataset_types, k=100),
    'hardware_used': random.choices(hardware_options, k=100),
    'num_layers': np.random.randint(1, 10, 100) # For deep learning models
}

df_ml_experiments = pd.DataFrame(data_ml_experiments)

# Display the first few rows of the generated DataFrame
print("Generated Unsupervised ML Experiments DataFrame head:")
display(df_ml_experiments.head())

# Save the DataFrame to a CSV file
ml_experiments_csv_file_path = 'unsupervised_ml_experiments_data.csv'
df_ml_experiments.to_csv(ml_experiments_csv_file_path, index=False)
print(f"\nUnsupervised ML experiments dataset saved to {ml_experiments_csv_file_path}")
```
