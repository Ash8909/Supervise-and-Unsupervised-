```python
import pandas as pd
import numpy as np

# Number of students
n_students = 65

# Generate student names (simple placeholders)
names = [f'Student_{i+1}' for i in range(n_students)]

# Generate ages between 18 and 25 (inclusive)
ages = np.random.randint(18, 26, n_students)

# Generate grades between 60 and 100 (inclusive)
grades = np.random.randint(60, 101, n_students)

# Create a dictionary from the generated data
data = {
    'Name': names,
    'Age': ages,
    'Grade': grades
}

# Create the pandas DataFrame
student_df = pd.DataFrame(data)

# Display the first few rows of the created dataset
display(student_df.head())

print(f"\nDataset created with {len(student_df)} students.")
```
